# Resuestas A Las Preguntas


a) ¿Qué significa que la desviación estándar es casi dos veces mayor que la MAD?

La desviación estándar (SD) mide la dispersión de los datos considerando todas las observaciones y es sensible a los valores atípicos. La mediana de las diferencias absolutas (MAD) mide la dispersión respecto a la mediana y es más resistente a outliers. Si la SD es aproximadamente el doble de la MAD, significa que los datos tienen una distribución con colas largas o presencia de valores atípicos que afectan la dispersión.


b) ¿Cómo puedo mejorar esta situación?

Para reducir la diferencia entre la SD y la MAD, se pueden tomar las siguientes medidas:

Eliminar o reducir el impacto de valores atípicos, mediante métodos como winsorización o transformación logarítmica.

Usar métricas más robustas, como la MAD en lugar de la SD en análisis donde los outliers afectan los resultados.

Aplicar normalización o transformación de datos para hacer la distribución más simétrica.

c) Explique el resultado obtenido al calcular el IQR (Rango Intercuartílico).

El IQR mide la dispersión de la parte central de los datos (Q3 - Q1) y excluye valores extremos. Si el IQR es pequeño, indica que la mayoría de los datos están concentrados cerca de la mediana. Un IQR grande sugiere que los datos tienen mayor variabilidad en la parte central de la distribución. Además, se usa para detectar valores atípicos mediante la regla de 1.5 veces el IQR.

d) ¿Qué información nos suministran los percentiles de tasa de homicidios por estados?

Los percentiles nos ayudan a entender la distribución de la tasa de homicidios:

25% de los estados tienen una tasa de homicidios menor a 2.425.

Estos valores nos muestran que la mayoría de los estados tienen tasas relativamente bajas, pero hay algunos con valores significativamente más altos.

e) ¿Qué información suministra el diagrama de caja con respecto a la tasa de homicidios por estado?

El diagrama de caja nos proporciona información clave:

**Mediana:** Se encuentra cerca de 4.0, lo que confirma que la mayoría de los estados tienen tasas de homicidio relativamente bajas.

**Rango intercuartílico (IQR):** Va desde 2.425 hasta 5.550, lo que indica que la mayoría de los estados tienen tasas en ese rango.

**Valores atípicos:** Si hay puntos fuera de los bigotes, representan estados con tasas de homicidios excepcionalmente altas.

# Análisis de otra dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import trim_mean

import pandas as pd

#Cargar dato desde google drive
from google.colab import drive
drive.mount("/content/drive")

df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/iris.csv")

# Mostrar las primeras filas para entender su estructura
df.head()

# Seleccionar columnas para análisis
col1 = df["Sepal.Length"]
col2 = df["Petal.Length"]

# Función para calcular estadísticos
def calcular_estadisticos(columna):
    estadisticos = {
        "Media": round(columna.mean(), 3),
        "Mediana": round(columna.median(), 3),
        "Desviación estándar": round(columna.std(), 3),
        "Varianza": round(columna.var(), 3),
        "Mínimo": round(columna.min(), 3),
        "Máximo": round(columna.max(), 3),
        "Rango": round(columna.max() - columna.min(), 3),
        "Q1 (25%)": round(columna.quantile(0.25), 3),
        "Q2 (50%) (Mediana)": round(columna.median(), 3),
        "Q3 (75%)": round(columna.quantile(0.75), 3),
        "IQR (Q3 - Q1)": round(columna.quantile(0.75) - columna.quantile(0.25), 3),
        "Trim Mean (10%)": round(trim_mean(columna, proportiontocut=0.1), 3)
    }
    return pd.Series(estadisticos)

# Crear tabla de resultados
tabla_estadisticos = pd.DataFrame({
    "Sepal.Length": calcular_estadisticos(col1),
    "Petal.Length": calcular_estadisticos(col2)
})

# Mostrar resultados
tabla_estadisticos

# Crear figura con dos gráficos separados (Boxplots)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Boxplot para Sepal.Length
sns.boxplot(y=col1, ax=axes[0], color="blue")
axes[0].set_title("Boxplot - Sepal Length")

# Boxplot para Petal.Length
sns.boxplot(y=col2, ax=axes[1], color="red")
axes[1].set_title("Boxplot - Petal Length")

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()

# Crear diccionarios con los valores de los estadísticos
estadisticos_col1 = {
    "Media": col1.mean(),
    "Mediana": col1.median(),
    "Desv. Est.": col1.std(),
    "Mínimo": col1.min(),
    "Máximo": col1.max(),
    "Q1 (25%)": col1.quantile(0.25),
    "Q3 (75%)": col1.quantile(0.75),
    "Trim Mean (10%)": trim_mean(col1, 0.1)
}

estadisticos_col2 = {
    "Media": col2.mean(),
    "Mediana": col2.median(),
    "Desv. Est.": col2.std(),
    "Mínimo": col2.min(),
    "Máximo": col2.max(),
    "Q1 (25%)": col2.quantile(0.25),
    "Q3 (75%)": col2.quantile(0.75),
    "Trim Mean (10%)": trim_mean(col2, 0.1)
}

# Crear figura con dos gráficos separados (Barras de estadísticos)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x=list(estadisticos_col1.keys()), y=list(estadisticos_col1.values()), ax=axes[0], color="blue")
axes[0].set_title("Estadísticos - Sepal Length")
axes[0].tick_params(axis='x', rotation=45)

sns.barplot(x=list(estadisticos_col2.keys()), y=list(estadisticos_col2.values()), ax=axes[1], color="red")
axes[1].set_title("Estadísticos - Petal Length")
axes[1].tick_params(axis='x', rotation=45)

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()

# Datos para la media y desviación estándar
media_col1 = col1.mean()
std_col1 = col1.std()
valores_col1 = [media_col1 - std_col1, media_col1, media_col1 + std_col1]
nombres_col1 = ["Media - Desv. Est.", "Media", "Media + Desv. Est."]

media_col2 = col2.mean()
std_col2 = col2.std()
valores_col2 = [media_col2 - std_col2, media_col2, media_col2 + std_col2]
nombres_col2 = ["Media - Desv. Est.", "Media", "Media + Desv. Est."]

# Crear figura con dos gráficos separados (Medias y Desviaciones)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x=nombres_col1, y=valores_col1, ax=axes[0], color="blue")
axes[0].set_title("Media y Desv. Est. - Sepal Length")
axes[0].tick_params(axis='x', rotation=45)

sns.barplot(x=nombres_col2, y=valores_col2, ax=axes[1], color="red")
axes[1].set_title("Media y Desv. Est. - Petal Length")
axes[1].tick_params(axis='x', rotation=45)

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()

# Crear gráficos de línea para ver evolución de valores
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
x_values = np.linspace(0, len(col1) - 1, len(col1))

# Graficar la primera columna
axes[0].plot(x_values, col1, label="Sepal Length", color="blue", alpha=0.7)
axes[0].set_title("Representación continua - Sepal Length")
axes[0].set_xlabel("Índice")
axes[0].set_ylabel("Valor")
axes[0].legend()
axes[0].grid(True)

# Graficar la segunda columna
axes[1].plot(x_values, col2, label="Petal Length", color="red", alpha=0.7)
axes[1].set_title("Representación continua - Petal Length")
axes[1].set_xlabel("Índice")
axes[1].set_ylabel("Valor")
axes[1].legend()
axes[1].grid(True)

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()
```
